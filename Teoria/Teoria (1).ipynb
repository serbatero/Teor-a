<p> 3. Dado un autómata finito determinístico $M$ , construya un nuevo autómata finito determinístico $M'$ con el mismo alfabeto de entrada y con la propiedad de que para todos los $u \in \sum$ , $u$ es aceptado por $M'$ si y solo si $u$ NO es aceptado por $M$. </p>

<h3>Solución</h3>

Existe un automata deterministico $M = (Q_M,\sum_M,\delta_M,q_M,F_M)$ Entonces contruimos un automata $M'$ con el mismo alfabeto de entrada, es decir que $\sum_M'=\sum_M$. Ademas que tiene la propiedad de que para todos los $u \in \sum$ , $u$ es aceptado por $M'$ si y solo si $u$ NO es aceptado por $M$, entonces quiere decir que todos los estados finales de $M'$ deben ser todos los estados que no son finales de $M$.

entonces $M' = (Q_{M'},\sum_{M'},\delta_{M'},q_{M'},F_{M'})$  donde $F_{M'} = Q - F_M$


<p>4. Dados dos autómatas finitos determinísticos $M_{1}$ y $M_{2}$, con el mismo alfabeto de entrada, construya un tercer autómata finito determinísticos ($M$) con la propiedad que $u \in \sum_{*}$ es aceptado por $M$ si y solo si es aceptado por ambos $M_{1}$ y $M_{2}$ </p>

<h3>Solución</h3>

tenemos que: 

<p style="text-align: center;">$M_1 = (Q_1,\sum_1,\delta_1,q_1,F_1)$<p/>
<p style="text-align: center;">$M_2 = (Q_2,\sum_2,\delta_2,q_2,F_2)$<p/>         
<p style="text-align: center;">$M = (Q,\sum,\delta,q,F)$<p/>              

Sabemos que el lenguaje de M va a ser igual a la interseccion entre el lenguaje de $M_1$ y el lenguaje de $M_2$, entonces definimos $M$ como:

$Q = Q_1 X Q_2 =${$(q^*,q')| q^* \in Q_1, p' \in Q_2$}

$\sum$ es el mismo de $M_1$ y $M_2$

$\delta((q^*,q'),\alpha)=\delta(q^*,\alpha),\delta(q',\alpha)$

$q_0 = (q_{01},q_{02})$

$F = F_1 X F_2$ = {$(q^*,q')| q^* \in F_1, p' \in F_2$}

<p>8. Pruebe o desapruebe la siguiente afirmación: El siguiente autómata acepta el lenguaje $L = ${$w|w = a^i b^j$ o $w = b^i c^j,   i \geq 1  j \geq 1$} 
<p/>
<img src="https://github.com/serbatero/Teoria/blob/master/Teoria/grafo.PNG?raw=true">



<p>Si acepta el lenguaje, ya que al ser una operación o, si cumple con ambas proposiciones $w = a^i b^j$ y $w = b^i c^j$ lo acepta, si cumple aunque sea con alguna de las dos, también lo acepta, así que sólo basta con probar una de ellas.
entonces sólo basta mirar los 3 primeros estados y ver que para $w = a^i b^j$ lo acepta ya que en el estado 1 es el inicio, en el estado 2 recibe $a$ como entrada, cumpliendose $i \geq 1$ para $a^i$ y por ultimo en el estado 3 recibe a $b$ como entrada, cumpliendose $j \geq 1$ para $b^j$, siendo este el estado final.<p/>
